```python
import pygame
import sys
import random

# Initialize pygame
pygame.init()

# Screen settings
WIDTH, HEIGHT = 800, 500
screen = pygame.display.set_mode((WIDTH, HEIGHT))
pygame.display.set_caption("Circle vs Square")

# Colors
WHITE = (255, 255, 255)
BLACK = (0, 0, 0)
RED = (220, 50, 50)
BLUE = (50, 100, 220)
GREEN = (50, 200, 50)

# Font
font = pygame.font.SysFont(None, 36)

# Clock
clock = pygame.time.Clock()

# Game state
circle_hp = 100
square_hp = 100

turn = "circle"
game_over = False
message = "Circle's turn! Press SPACE to attack."

# Positions
circle_pos = (200, 250)
square_rect = pygame.Rect(550, 200, 100, 100)

# Main game loop
running = True

while running:

    screen.fill(WHITE)

    # --- Events ---
    for event in pygame.event.get():

        if event.type == pygame.QUIT:
            running = False

        if event.type == pygame.KEYDOWN and not game_over:

            # SPACE = attack
            if event.key == pygame.K_SPACE:

                damage = random.randint(8, 20)

                if turn == "circle":
                    square_hp -= damage
                    message = f"Circle attacks for {damage} damage!"
                    turn = "square"

                else:
                    circle_hp -= damage
                    message = f"Square attacks for {damage} damage!"
                    turn = "circle"

                # Clamp HP
                circle_hp = max(circle_hp, 0)
                square_hp = max(square_hp, 0)

                # Win conditions
                if circle_hp <= 0:
                    message = "Square wins!"
                    game_over = True

                elif square_hp <= 0:
                    message = "Circle wins!"
                    game_over = True

                else:
                    message += f" Next turn: {turn}"

    # --- Draw circle ---
    pygame.draw.circle(screen, BLUE, circle_pos, 50)

    # --- Draw square ---
    pygame.draw.rect(screen, RED, square_rect)

    # --- HP Text ---
    circle_text = font.render(f"Circle HP: {circle_hp}", True, BLACK)
    square_text = font.render(f"Square HP: {square_hp}", True, BLACK)

    screen.blit(circle_text, (50, 50))
    screen.blit(square_text, (500, 50))

    # --- Message Text ---
    msg_surface = font.render(message, True, GREEN)
    screen.blit(msg_surface, (50, 430))

    # Update display
    pygame.display.flip()

    # FPS
    clock.tick(60)

pygame.quit()
sys.exit()
```
